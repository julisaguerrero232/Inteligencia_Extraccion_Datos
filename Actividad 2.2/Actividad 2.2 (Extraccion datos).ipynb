```python
#importamos librerias requeridas
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .xlsx sin indice
df = pd.read_excel('Datos de Facturación.xlsx')
df
```

```python
#Verificamos informacion de dataframe
df.info()
```

```python
# filtrar(valores: 1000 a 2000) 
df['CVE_CLPV'] = pd.to_numeric(df['CVE_CLPV'], errors='coerce') # Convierte los valores de la columna CVE_CLPV a datos numéricos
filtro1 = df[df['CVE_CLPV'].between(1000, 2000)] #muestra solo las filas donde el valor de CVE_CLPV está entre 1000 y 2000, inclusivo.
filtro1
filtro1.to_csv("Entregable1.csv")
```

```python
#para excluir aquellas en las que la columna CVE_VEND contiene los valores 4 o 5
filtro2 = df[~df["CVE_VEND"].isin([4, 5])] #Devuelve una serie booleana (True/False) indicando si cada valor de la columna está en la lista [4, 5]
filtro2
filtro2.to_csv("Entregable2.csv")
```

```python
# Convertir 'FECHA_ENT' a tipo datetime sin especificar un formato
df['FECHA_ENT'] = pd.to_datetime(df['FECHA_ENT'], errors='coerce')

# Filtrar solo las filas con la fecha '2022-02-28' (sin la hora)
df['FECHA_ENT'] = df['FECHA_ENT'].dt.date
filtro3 = df[df['FECHA_ENT'] == pd.to_datetime('2022-02-28').date()]

# Verificar el resultado
filtro3

filtro3.to_csv("Entregable3.csv")
```

```python
# Filtrar las filas donde CAN_TOT < 5951.7 o STATUS == "E"
filtro4 = df[(df['CAN_TOT'] < 5951.7) | (df['STATUS'] == 'E')] #Combina ambas condiciones, devolviendo las filas que cumplen al menos una de las condiciones
filtro4

filtro4.to_csv("Entregable4.csv")
```

```python
#Solo las columnas: CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT
filtro6 = df[['CVE_DOC', 'FECHA_ENT', 'FECHA_VEN', 'CAN_TOT']]
filtro6
filtro6.to_csv("Entregable6.csv")
```

```python
# Seleccionar solo las filas de 7001 a 7099
filtro7 = df.iloc[7001:7099]
filtro7
filtro7.to_csv("Entregable7.csv")
```
